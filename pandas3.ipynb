```python
import numpy as np 
import pandas as pd 
```

```python
np.random.seed(20)
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2023-07-08',periods=10),columns=['a','b','c','d'])
df
```

```python
df['new_square_root_d']=df['d'].apply(lambda x:x**.5)
```

```python
df['new_apply_log']=df['d'].apply(lambda x:np.log(x))
```

```python
df
```

```python
## if we don't need nan value than we can do this 
def test(x):
    if x<0:
        return x
    else:
        return np.log(x)
```

```python
df['new_apply_log']=df['d'].apply(test)
```

```python
def test1(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
df['test_for_neg_or_pos']=df['a'].apply(test1)
```

```python

def wrap(string, max_width):
    a=''
    j=0
    for i in string:
        j=j+1
        if j<=max_width:
            a=a+i
        else:
            print(a)
            a=''
            j=0
    return

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```
